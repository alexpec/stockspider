```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
companies_codes = [
    'PETR', # Petrobras
    'ABEV', # Ambev
    'CMIG', # CMIG
    'GGBR', # Gerdau
    'ITUB', # Itaú
    'BBDC', # Bradesco
    'BBAS', # Banco do Brasil
    'VALE', # Vale
    'ALPA', # Alpargatas
    'ALUP', # Alupar
    'AMAR', # Mariza
    'BIDI', # Banco Inter
    'BOBR', # Bombril
    'BRKM', # Brasken
    'BRML', # Br Malls
    'CNTO', # Centauro
    'CVCB', # CVC
    'ELEK', # Eletrobras
    'ELET', # Eletropaulo
    'ENGI', # Energisa
    'LAME', # Lojas Americanas
    'LIPR', # Eletropar
    'OGXP', # OGX Petroleo
    'NATU', # Natura
    'WHRL', # Whirlpool
    
]
companies_codes.sort()

stock_codes = {}
stocks_list = []

codes_processed_file = '../Python/processed_codes.csv'
file = open(codes_processed_file, 'r')
line = file.readline()

while line:
    line = line.strip()
    id, name = line.split(',')
    stocks_list.append(id)
    line = file.readline()

file.close()

for i in companies_codes:
    stock_codes[i] = [s for s in stocks_list if i in s]
    
```

```python
timeseries_file = '../Python/outputdata.csv'
df = pd.read_csv(timeseries_file)
df['Date'] = pd.to_datetime(df['Date'], format='%d/%m/%Y')
df
```

```python
blue_chips = sum(stock_codes.values(), [])
print(blue_chips)
```

```python
df['Movement'] = df['Price Close'] - df['Price Open']
df
```

```python
bc_df = df.loc[(df['Stock Code'].isin(blue_chips))].copy() #& (df['Date'] < '2019-06-01')].copy()
bc_df
```

```python
df_pivot = pd.pivot_table(bc_df, index=[bc_df['Date']], columns=bc_df['Stock Code'], values=['Movement', 'Volume'])
df_pivot
```

```python
#Valid M-Monthly, 2M-Two Months, ...
#This code returns the date as a column again

#data = df_pivot.resample('M').sum().reset_index().copy() 
```

```python
df_pivot = df_pivot.resample('D').sum()
df_pivot
```

```python
movements = df_pivot.to_numpy().transpose()
```

```python
from sklearn.preprocessing import Normalizer, StandardScaler
norm = Normalizer()
stdm = StandardScaler()
```

```python
from sklearn.pipeline import make_pipeline
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters =5, max_iter = 5000)
pipeline = make_pipeline(norm, kmeans)
pipeline.fit(movements)
labels = pipeline.predict(movements)
```

```python
df1 = pd.DataFrame({'labels':labels, 'companies':[i[1] for i in df_pivot.columns.to_list()]}).sort_values(by=['labels'], axis=0)
df1
```
